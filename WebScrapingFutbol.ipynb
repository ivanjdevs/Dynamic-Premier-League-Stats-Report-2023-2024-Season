```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd # tha famous scientific library to deal with dataframes
import html5lib  #  the parser we want to use to parse the HTML content
```

<p><font size = 3> Download the webpage at the url </font></p>

```python
url = "https://fbref.com/es/comps/9/Estadisticas-de-Premier-League"

response = requests.get(url)
print(response.status_code)
#If we get status code=200, the connection is ok
```

```python
#The .text method extracts the HTML content as a string in order to make it readable.
data  = requests.get(url).text
## print(data) Comenté esta instrucción porque ya se verificó que efectivamente se leyó la página
```

#### 2. Parse the html content

We will take the raw HTML content of a webpage or a string of HTML code which needs to be parsed and transformed into a structured, hierarchical format that can be more easily analyzed and manipulated in Python. This can be done using a Python library called Beautiful Soup.

```python
#Create a soup object
sopaesta=BeautifulSoup(data, "html5lib")
```

```python
#Leer todas las tablas HTML y guardarlas en un objeto pandas
table_list = pd.read_html(url, flavor='bs4')
# ¿Cuántas tablas hay?
len(table_list)
```

```python
## De la página web solo vamos a usar las dos primeras tablas. La primera ha quedado guardada en en table_list en el índice 0
Positions=table_list[0]
Positions
```

```python
## Eliminando la column RL y Notas
Positions.drop(['RL','Notas'], axis=1, inplace=True)
Positions
```

<h4>Leyendo ahora la tabla 2 que quedó en el objeto table list </h4>

```python
TeamStats = table_list[2]
TeamStats
```

#### Vemos que tenemos como varios niveles en los encabezados de las columnas, una vaina fea ahí. Eliminemos el nivel de mas arriba.

```python
TeamStats.columns=TeamStats.columns.droplevel()
TeamStats
```

```python
#Veamos los nombres de las columnas
print(TeamStats.columns)
```

```python
##Eliminando unas columnas que no usaremos
TeamStats.drop(['PL','xG', 'npxG', 'xAG','npxG+xAG', 'PrgC', 'PrgP', 'Gls.', 'Ast', 'G+A', 'G-TP', 'G+A-TP','xG', 'xAG', 'xG+xAG', 'npxG', 'npxG+xAG'], axis=1, inplace=True)
TeamStats
```

```python
##Ejecutando merge para fusionar los dos dataframes Position and TeamStats. Le decimos las columnas por las que queremos se una (las que se repiten)
GenStats=pd.merge(Positions, TeamStats, how ='left', on=['Equipo','PJ'])

#Ordenando el dataset por orden alfabético de acuerdo con la columna Equipo; inplace=True para que haga la ordenación en el mismo dataframe y no en una copia,
# y finalmente ignore_index=True para que descarte el cambio de orden que ocurrirá por cuenta de ejecutar 'sort_values' en la columna 'indices'  y asigne uno nuevo.
GenStats.sort_values('Equipo', inplace=True, ignore_index=True)

## Visualizar como quedó el dataframe 
GenStats

##Otra forma de unir los dos dataframes, solo qu esta dejará la columna usada para hacer la concatenación (en este caso Equipo) en un nivel y el resto de headers en otro
#pru=pd.concat([Positions.set_index('Equipo'),TeamStats.set_index('Equipo')], axis=1, join='outer')
#pru
```
